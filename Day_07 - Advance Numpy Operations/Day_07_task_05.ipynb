### Program 5: Aggregation Functions on 2D Arrays

```python
import numpy as np

marks = np.array([
    [85, 90, 88],
    [78, 82, 80],
    [92, 95, 94]
])

print("Total Marks (All):", np.sum(marks))
print("Column-wise Sum:", np.sum(marks, axis=0))
print("Row-wise Average:", np.mean(marks, axis=1))
print("Maximum Marks:", np.max(marks))
print("Minimum Marks:", np.min(marks))
```
