### Program 1: Advanced Indexing (Boolean & Fancy Indexing)

```python
import numpy as np

arr = np.array([10, 25, 30, 45, 50, 65])

# Boolean indexing
greater_than_30 = arr[arr > 30]

# Fancy indexing
selected_elements = arr[[1, 3, 5]]

print("Original Array:", arr)
print("Elements > 30:", greater_than_30)
print("Selected Elements:", selected_elements)
```
